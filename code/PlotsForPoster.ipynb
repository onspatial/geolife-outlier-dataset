```python
import pandas
import matplotlib.pyplot as plt
from utils.GeoDataFrame import getCleanGeoDataFrame
```

# GeoLife Dataset

```python
inputRawFileName = 'raw_data.tsv'
geoLifeDataFrame = getCleanGeoDataFrame( inputRawFileName, '../data/geolife', limit=0)
```

```python
workingGeoLifeDataFrame = geoLifeDataFrame.copy()
print(workingGeoLifeDataFrame.head())

workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split(',').str[0]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split(' ').str[1]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split(':').str[2]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split('-').str[0] + '-' + workingGeoLifeDataFrame['ArrivingTime'].str.split ('-').str[1] + workingGeoLifeDataFrame['ArrivingTime'].str.split ('-').str[2]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split('-').str[0] + '-' + workingGeoLifeDataFrame['ArrivingTime'].str.split ('-').str[1] + '-' + workingGeoLifeDataFrame['ArrivingTime'].str.split ('-').str[2]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split(':').str[0] + ':' + workingGeoLifeDataFrame['ArrivingTime'].str.split (':').str[2]
# workingGeoLifeDataFrame['ArrivingTime'] = workingGeoLifeDataFrame['ArrivingTime'].str.split(':').str[1] + ':' + workingGeoLifeDataFrame['ArrivingTime'].str.split (':').str[2]
print(workingGeoLifeDataFrame.head())

```

```python
workingGeoLifeDataFrame = workingGeoLifeDataFrame.groupby(['ArrivingTime']).size().reset_index(name='counts')
workingGeoLifeDataFrame['ArrivingTime'] = pandas.to_datetime(workingGeoLifeDataFrame['ArrivingTime'])
maxCount = workingGeoLifeDataFrame['counts'].max()
print('maxCount: ', maxCount)

dataOfTheMaxCount = workingGeoLifeDataFrame.loc[workingGeoLifeDataFrame['counts'] == maxCount]
print('dataOfTheMaxCount: ', dataOfTheMaxCount)
print(workingGeoLifeDataFrame.head())
```

```python
plt.figure(figsize=(15, 5))  # Set the size of the figure
plt.plot(workingGeoLifeDataFrame['ArrivingTime'], workingGeoLifeDataFrame['counts'], marker='o')  # Create a line plot with markers

# Customize the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.xlabel('Time')
plt.ylabel('Stay Points Counts')
plt.grid(True)  # Add a grid for easier reading

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python

plt.figure(figsize=(15, 5))  # Set the size of the figure
plt.scatter(workingGeoLifeDataFrame['ArrivingTime'], workingGeoLifeDataFrame['counts'])  # Create a scatter plot

# Customize the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.xlabel('Time')
plt.ylabel('Stay Points Counts')
plt.grid(True)  # Add a grid for easier reading

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python

plt.figure(figsize=(15, 5))  # Set the size of the figure
# Create a scatter plot with larger dots by setting the `s` parameter
plt.scatter(workingGeoLifeDataFrame['ArrivingTime'], workingGeoLifeDataFrame['counts'], s=3)  # You can change 50 to any other value as per your preference for dot size

# Customize the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.xlabel('Time')
plt.ylabel('Stay Points Counts')
plt.grid(True)  # Add a grid for easier reading

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import pandas
dataFrame = pandas.read_csv('../data/pol/pol-traj-2020-02-01.tsv', sep='\t', header=None)
## remove the fist row
dataFrame = dataFrame.iloc[1:]

dataFrame[1] = dataFrame[1].str.split(' ').str[1] + ',' + dataFrame[1].str.split(' ').str[2]
# remove '(' and ')' from the location tuple
dataFrame['SimulationTime'] = dataFrame[0]
dataFrame[1] = dataFrame[1].str.replace('(', '')
dataFrame[1] = dataFrame[1].str.replace(')', '')
dataFrame['Latitude'] = dataFrame[1].str.split(',').str[1]
dataFrame['Longitude'] = dataFrame[1].str.split(',').str[0]
dataFrame['AgentID'] = dataFrame[2]
#remove extra columns
dataFrame.drop(0, axis=1, inplace=True)
dataFrame.drop(1, axis=1, inplace=True)
dataFrame.drop(2, axis=1, inplace=True)

dataFrame.reset_index(drop=True, inplace=True)
print(dataFrame.head())
```

```python

# save to csv
# headers:
# SimulationTime, Latitude, Longitude, AgentID

dataFrame.columns = ['SimulationTime', 'Latitude', 'Longitude', 'AgentID']
dataFrame.to_csv('../data/pol/pol-cleaned-traj-2020-02-01.tsv', sep=' ', index=False, header=True)
```
